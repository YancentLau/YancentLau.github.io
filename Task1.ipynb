# For Loop

As you see, your agent robot will move a step forward by command 
```
agent.move("forward")
```
When you need to move few more step, it is better to use for loop to instead of copying a same code.

```
for i in range (3):
   agent.move("forward")
```
What we have here
- <b>for (variable) in range (looping time):</b> the variable will loop inside the range that you set. 
- Let's talk about the previous example
- &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; it means code inside for loop will loop for <b>3</b> times
- &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; </b>i</b> will become 0 in the first loop
- &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; </b>i</b> will become 1 in the second loop
- &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; </b>i</b> will become 2 in the third loop
- So this for loop will order your agent robot to move forward by three times.
