# Agent Say

But here is a problem, eventhough your agent know what material that is, you still cannot know it.

So you need to command your agent to say by :
```

```
You can put the message that you want your agent to say, then put the message inside the bracket. For example :


````python
agent.say("Halo")
````


## If - Else
In python, there are a simple logic, which is if else.


````python
if CONDITION :
    DO IF CONDITION IS TRUE
else:
    DO IF CONDITION IS FALSE
````

Example :

````python
if agent.inspect("down")=="gold_block":
    agent.move("forward")
else:
    agent.say("Hi")
````

In this example
```
 Line 1 :      if the result of agent detect "down" is EQUAL TO "gold_block" 
 Line 2 :      when true (i.e. result of detection is "gold_block"), it will do   =>   agent.move("forward")
 Line 3 :      else :
 Line 4 :      when false (i.e. result of detection is not "gold_block") , it will do   =>   agent.say("Hi")
```

## Logical Conditions
