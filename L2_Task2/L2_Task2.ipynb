# While loop

Python has two loop commands :
```
1.  for loops
2.  while loops
```

We tried to use for loops in the level 1. Now we can try while loops.



````python
while CONDITION:
    DO IF THE RESULT OF CONDITION IS TRUE
````

### Logical Conditions
In programming, we always use some logical conditions which come from mathematics:
```
    MEANINGS                   EXAMPLES
    -----------------------------------
1.  Equals                       a == b
2.  Not equals                   a != b
3.  Less than                    a <  b
4.  Less than or Equal to        a <= b
5.  Greater than                 a >  b
6.  Greater than or Equal to     a >= b

## Example 1

````python
while True:
    agent.move("forward")
    break 
````

Here is the explanation of the example 1:

```
Line 1    :  while CONDITION  
             (Since we already set the condition as TRUE, so that the program will keep looping the code inside this while loop)
Line 2    :  when true , it will keep doing =>  agent.move("forward")
Line 3    :                                 =>  break 
                                                this command will break the while loop
```

## Example 2


````python
while agent.inspect("down")!="smooth_stone":
    agent.move("forward")
    agent.say("yes")
agent.say("no")
````

Here is the explanation of the example:
```
Line 1    :  while CONDITION (i.e.the result of agent detect "down" is not a "smooth_stone")
Line 2    :  when true (i.e. result of detection is not "smooth_stone"), it will keep doing =>   agent.move("forward")
Line 3    :                                                                                   =>   agent.say("yes")
Line 4    :  once the CONDITION is false, the program will break out the while loop and run   =>   agent.say("no")
