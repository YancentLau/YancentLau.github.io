## If - Else
In python, there are a simple logic, which is if else.


````python
if CONDITION :
    DO IF CONDITION IS TRUE
else:
    DO IF CONDITION IS FALSE
````

Example :

````python
if agent.inspect("down")=="gold_block":
    agent.move("forward")
else:
    agent.say("Hi")
````

In this example
```
 Line 1 :      if the result of agent detect "down" is EQUAL TO "gold_block" 
 Line 2 :      when true (i.e. result of detection is "gold_block"), it will do   =>   agent.move("forward")
 Line 3 :      else :
 Line 4 :      when false (i.e. result of detection is not "gold_block") , it will do   =>   agent.say("Hi")
```

# While true

Python has two loop commands :
```
1.  for loops
2.  while loops
```

We tried to use for loops in the level 1. Now we can try while loops.



````python
while CONDITION:
    DO IF CONDITION IS TRUE
````

````python
while agent.inspect("down")!="smooth_stone":
    agent.move("forward")
    agent.say("yes")
agent.say("no")
````

Here is the explanation of the example:
```
Line 1    :  while CONDITION (i.e.the result of agent detect "down" is not a "smooth_stone")
Line 2    :  when true (i.e. result of detection is not "smooth_stone"), it will keep doing =>   agent.move("forward")
Line 3    :                                                                                   =>   agent.say("yes")
Line 4    :  once the CONDITION is false, the program will break out the while loop and run   =>   agent.say("no")
